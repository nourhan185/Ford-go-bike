## Investigation Overview

> In this analysis I would like to figur out how trip duration is dependent on other specifications from the dataset like user type , gender and age .

## Dataset Overview

> There are 174952 fordgobike trips in the dataset with 16 specifications (duration_sec, start_time, end_time, start_station_id, start_station_name,day_of_month,trip,hour_of_day, end_station_id, end_station_name, bike_id, user_type, member_birth_year, member_gender, age,bike_share_for_all_trip).

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import nbconvert 
%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
df = pd.read_csv("ford go bike.csv",parse_dates=['start_time','end_time'])
```

## (The most common 10 trips)

> The most common ten trips are(Berry st to sanfrancisco /grande Ave to part station/ sanfrancisco to the embarcadero/ Bart station to grande Ave/ the embarcadero to steuart st / townnsend st to san francisco/ san fernando to 5th st / market st to Bart station/ 5th st to san fernando / market st to san francisco)
Also we have noticed that some stations are linked to another stations and switches between them back and forth in the trip like : grane Ave - Bart statin & san fernando - 5th st

```python
df['trip']=df['start_station_name']+':'+ df['end_station_name']
freq=df["trip"].value_counts()
trip_order= freq.index[:10]
base_color=sb.color_palette()[6]
sb.countplot(data= df, y= "trip",color= base_color, order=trip_order);
plt.xticks(rotation=90);
plt.title('The most common 10 trips');
```

## ( The most commom  hours for starting)

> The most commom hours of day for starting are : in morning from 8:9 and evening from 4:6.

```python
df['hour_of_day']=df.start_time.dt.hour
freq=df["hour_of_day"].value_counts()
hours_order= freq.index[:5]
base_color=sb.color_palette()[2]
sb.countplot(data= df, x= "hour_of_day",color= base_color, order=hours_order);
plt.title('The most commom hours for starting');
```

## (The correlation between trip duration and age)

#### There is no correlation between the two variables .



```python
df['age']=2021-df['member_birth_year']
sb.regplot(data = df, x = 'age', y = 'duration_sec', truncate=False, x_jitter=0.3, scatter_kws={'alpha':1/20})
plt.title('the correlation between trip duration and age')
plt.xlabel('Age (year)')
plt.ylabel('Duration(sec)');
```
